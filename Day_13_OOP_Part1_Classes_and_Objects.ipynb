## 1. What is OOP?

**Object-Oriented Programming** organises code around **objects** — bundles of **data**
(variables) and **behaviour** (functions) that belong together — instead of loose functions.

| Procedural style | Object-Oriented style |
|---|---|
| Data and functions are separate | Data + functions bundled into objects |
| Harder to model real things | Naturally models real-world things |
| Reuse via functions | Reuse via classes & inheritance |

### The Four Pillars of OOP
1. **Encapsulation** — bundle data + methods, hide internals *(Part 2)*.
2. **Inheritance** — a class reuses another class's features *(Part 2)*.
3. **Polymorphism** — same interface, different behaviour *(Part 3)*.
4. **Abstraction** — expose only the essentials, hide the rest *(Part 3)*.

This notebook covers the **foundation**: classes, objects, methods and variables.

## 2. Class vs Object

- **Class** — a **blueprint / template**. It describes what data and methods objects will have.
- **Object** — an **instance** of a class; a real thing built from the blueprint.

```text
        CLASS (blueprint)                OBJECTS (instances)
      +------------------+            +---------+   +---------+
      |      Car         |  ------->  |  car1   |   |  car2   |
      |  brand, color    |            | Toyota  |   |  BMW    |
      |  start(), stop() |            |  red    |   |  black  |
      +------------------+            +---------+   +---------+
```

**Key Notes:**
- One class → many objects, each with its **own** data.
- "Blueprint of a house" (class) vs "actual houses built" (objects).

## 3. Defining a Class & Creating Objects

**Syntax**
```python
class ClassName:
    # attributes and methods
    ...

obj = ClassName()   # create an object (instance)
```

**Key Notes:**
- Class names use **PascalCase** (e.g. `BankAccount`).
- Creating an object is called **instantiation**.

````python
class Dog:
    sound = "Woof"          # a simple class attribute

# Create two objects from the same blueprint
d1 = Dog()
d2 = Dog()

print(type(d1))             # <class '__main__.Dog'>
print(d1.sound, d2.sound)   # both share the class attribute
````

## 4. The `__init__` Constructor and `self`

`__init__` is a special method that runs **automatically** when an object is created. It is used
to set up the object's starting data. **`self`** refers to *the object being created/used*.

**Syntax**
```python
class Person:
    def __init__(self, name, age):
        self.name = name    # store data ON the object
        self.age = age
```

**Key Notes:**
- `self` is always the **first parameter** of instance methods — Python passes it automatically.
- You do **not** pass `self` yourself: `Person("Alice", 30)` — Python fills `self` in.

````python
class Person:
    def __init__(self, name, age):
        self.name = name        # instance variable
        self.age = age

# self is filled in automatically; we pass name and age
p1 = Person("Alice", 30)
p2 = Person("Bob", 25)

print(p1.name, p1.age)
print(p2.name, p2.age)
````

## 5. Instance Variables

**Instance variables** belong to **one specific object**. They are created with `self.` and each
object keeps its **own copy**.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Alice", 30)
p2 = Person("Bob", 25)

p1.age = 31             # changes ONLY p1
print(p1.age)           # 31
print(p2.age)           # 25 (unaffected)
````

## 6. Class Variables

**Class variables** are **shared by every object** of the class. They are defined directly inside
the class body (not inside `__init__`). Great for constants or counters.

**Key Notes — what about "static variables"?**
- In Python, a **class variable *is* the "static variable"** you may know from C++/Java — a single
  copy shared across all objects of the class.
- Python has **no separate `static` variable** keyword or type. So "class variable" and "static
  variable" describe the **same thing** here; Python simply calls it a *class variable*.
- (Don't confuse this with a **static *method*** in §10 — that's a method, not a variable.)

````python
class Person:
    number_of_people = 0            # class variable (shared)

    def __init__(self, name):
        self.name = name            # instance variable (per object)
        Person.number_of_people += 1  # update the shared counter

a = Person("Alice")
b = Person("Bob")
c = Person("Ravi")

print("total people:", Person.number_of_people)   # 3 (shared across all)
````

## 7. Instance vs Class Variables

| | Instance variable | Class variable |
|---|---|---|
| Defined | inside `__init__` with `self.` | in the class body |
| Belongs to | one object | the class (shared) |
| Copies | one per object | a single shared copy |
| Access | `obj.var` | `Class.var` or `obj.var` |
| Use for | per-object data (name, age) | shared data (counter, constant) |

> **Static variable?** Python has only these **two** kinds of variables. What other languages
> (C++/Java) call a *static variable* is exactly the **class variable** column above.

## 8. Instance Methods

An **instance method** works on a specific object's data. Its first parameter is **`self`**, so it
can read/change that object's instance variables.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):                          # instance method
        return f"Hi, I'm {self.name}, age {self.age}."

p = Person("Alice", 30)
print(p.greet())          # object.method() -> self is p automatically
````

## 9. Class Methods (`@classmethod`)

A **class method** works on the **class itself**, not one object. It takes **`cls`** (the class) as
its first parameter and is marked with **`@classmethod`**. Useful for data shared across all
objects (like our counter) or alternative constructors.

````python
class Person:
    number_of_people = 0

    def __init__(self, name):
        self.name = name
        Person.number_of_people += 1

    @classmethod
    def get_count(cls):               # cls is the class (Person)
        return cls.number_of_people

Person("Alice")
Person("Bob")
print(Person.get_count())             # 2  (called on the class)
````

## 10. Static Methods (`@staticmethod`)

A **static method** takes **neither `self` nor `cls`**. It is a plain helper function that lives
inside the class only because it *belongs there logically*. Marked with **`@staticmethod`**.

````python
class Person:
    @staticmethod
    def is_adult(age):                 # no self, no cls
        return age >= 18

# Can be called on the class OR an object
print(Person.is_adult(30))    # True
print(Person.is_adult(16))    # False
````

## 11. Instance vs Class vs Static Methods

| Method type | First parameter | Decorator | Can access |
|---|---|---|---|
| **Instance** | `self` | (none) | object data + class data |
| **Class** | `cls` | `@classmethod` | class data only |
| **Static** | — (none) | `@staticmethod` | nothing automatic (a plain helper) |

**Key Notes:**
- Use **instance** methods for per-object behaviour.
- Use **class** methods for class-wide data / alternative constructors.
- Use **static** methods for utility logic that fits the class but needs no object.

## 12. All Three Together (Complete Example)

This single `Person` class uses an instance method, a class method and a static method — the
classic reference example.

````python
class Person:
    number_of_people = 0                       # class variable

    def __init__(self, name, age):
        self.name = name                       # instance variables
        self.age = age
        Person.number_of_people += 1

    def greet(self):                           # instance method
        return f"Hello, my name is {self.name} and I am {self.age} years old."

    @classmethod
    def get_number_of_people(cls):             # class method
        return cls.number_of_people

    @staticmethod
    def is_adult(age):                         # static method
        return age >= 18

person1 = Person("Alice", 30)
person2 = Person("Bob", 25)

print(person1.greet())                 # instance method
print(Person.get_number_of_people())   # class method -> 2
print(Person.is_adult(30))             # static method -> True
print(Person.is_adult(16))             # static method -> False
````

## 13. Magic (Dunder) Methods

**Magic methods** (a.k.a. **dunder** = *double underscore*) let your objects work with built-in
Python syntax — printing, `+`, `==`, `len()`, indexing, etc. Python calls them **automatically**.

| You write | Python calls |
|---|---|
| `str(obj)` / `print(obj)` | `obj.__str__()` |
| `repr(obj)` | `obj.__repr__()` |
| `a + b` | `a.__add__(b)` |
| `a == b` | `a.__eq__(b)` |
| `len(obj)` | `obj.__len__()` |
| `obj[i]` | `obj.__getitem__(i)` |
| `item in obj` | `obj.__contains__(item)` |
| `obj()` | `obj.__call__()` |

## 14. `__str__` vs `__repr__`

- **`__str__`** → friendly text for **users** (used by `print()` and `str()`).
- **`__repr__`** → unambiguous text for **developers** (used by `repr()` and the interactive shell).

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"      # for users

    def __repr__(self):
        return f"Person('{self.name}', {self.age})"       # for developers

p = Person("Alice", 30)
print(p)          # uses __str__  -> Alice, 30 years old
print(str(p))     # __str__
print(repr(p))    # __repr__      -> Person('Alice', 30)
````

## 15. `__len__` — Make `len()` Work

Define `__len__` so Python's built-in `len()` works on your object.

````python
class Team:
    def __init__(self, members):
        self.members = members

    def __len__(self):
        return len(self.members)

t = Team(["Alice", "Bob", "Ravi"])
print(len(t))        # 3  -> calls t.__len__()
````

## 16. `__eq__`, `__lt__`, `__gt__` — Comparisons

Define these to make `==`, `<`, `>` work between your objects with your own logic.

````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks      # equal marks -> equal

    def __lt__(self, other):
        return self.marks < other.marks       # compare by marks

a = Student("Alice", 90)
b = Student("Bob", 75)

print(a == b)     # False  -> __eq__
print(a > b)      # True   -> __gt__ derived from __lt__
print(b < a)      # True   -> __lt__
````

## 17. `__add__` — Operator Overloading

Define `__add__` so the `+` operator works on your objects. (More on this in Part 3.)

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v = Vector(2, 3) + Vector(3, 5)     # calls __add__
print(v)                            # Vector(5, 8)
````

## 18. `__getitem__` / `__setitem__` — Indexing

Define these to let your object be indexed like a list or dict (`obj[i]`, `obj[i] = x`).

````python
class MyList:
    def __init__(self):
        self.items = [10, 20, 30]

    def __getitem__(self, index):        # obj[index]
        return self.items[index]

    def __setitem__(self, index, value): # obj[index] = value
        self.items[index] = value

m = MyList()
print(m[0])          # 10  -> __getitem__
m[0] = 99            # __setitem__
print(m[0])          # 99
````

## 19. `__call__` — Make an Object Callable

Define `__call__` to use an object **like a function**: `obj(...)`.

````python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, x):
        return x * self.factor

double = Multiplier(2)
print(double(10))     # 20  -> calls double.__call__(10)
print(double(7))      # 14
````

## 20. Special Built-in Dunder **Variables**

Besides dunder *methods*, Python defines special dunder **variables** that carry metadata.

| Variable | Meaning |
|---|---|
| `__name__` | Module name (`"__main__"` if run directly); a function's name as a string |
| `__doc__` | The docstring of a module/class/function |
| `__dict__` | Namespace dict of an object/class (its attributes) |
| `__class__` | The class an instance belongs to |
| `__bases__` | Tuple of a class's base (parent) classes |
| `__mro__` | Method Resolution Order of a class |
| `__module__` | Module in which a class was defined |

````python
class Person:
    "A simple person class."          # this becomes __doc__
    def __init__(self, name):
        self.name = name

p = Person("Alice")
print("__doc__  :", Person.__doc__)
print("__dict__ :", p.__dict__)        # {'name': 'Alice'}
print("__class__:", p.__class__.__name__)
print("__bases__:", Person.__bases__)  # (<class 'object'>,)

def greet():
    "says hello"
    return "hi"
print("function __name__:", greet.__name__)
print("function __doc__ :", greet.__doc__)
````

## 21. Common Mistakes

**1. Forgetting `self`**
```python
class A:
    def show():           # WRONG - missing self
        print("hi")
# RIGHT
class A:
    def show(self):
        print("hi")
```

**2. Using a class variable where you meant an instance variable** — a shared mutable class
variable (e.g. a list) is accidentally shared by *all* objects. Put per-object data in `__init__`.

**3. Passing `self` manually** — write `p.greet()`, not `p.greet(p)`.

**4. Confusing the class and the object** — `Person` is the blueprint; `Person("Alice", 30)` is
an object built from it.

## 22. Mini Examples

````python
# BankAccount: instance data + methods
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient funds"
        self.balance -= amount
        return self.balance

acc = BankAccount("Alice", 1000)
print(acc.deposit(500))    # 1500
print(acc.withdraw(2000))  # Insufficient funds
print(acc.withdraw(300))   # 1200
````

````python
# Circle: methods computing from instance data
class Circle:
    PI = 3.14159                 # class variable (constant)

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return Circle.PI * self.radius ** 2

    def circumference(self):
        return 2 * Circle.PI * self.radius

c = Circle(5)
print("area         :", round(c.area(), 2))
print("circumference:", round(c.circumference(), 2))
````

## 23. Summary

- **Magic/dunder methods** (`__str__`, `__eq__`, `__add__`, `__len__`, `__getitem__`, `__call__`)
